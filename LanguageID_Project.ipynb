```python
import pandas as pd
from datasets import load_dataset

print("Downloading dataset from Hugging Face... (This might take a few seconds)")

# 1. Load the dataset
dataset = load_dataset("papluca/language-identification")

# 2. Convert the training data into a Pandas DataFrame
df = pd.DataFrame(dataset['train'])

# 3. Filter for exactly 500 English rows and 500 Swahili rows to ensure a BALANCED dataset
df_english = df[df['labels'] == 'en'].head(500)
df_swahili = df[df['labels'] == 'sw'].head(500)

# 4. Combine them together
df_combined = pd.concat([df_english, df_swahili])

# 5. Rename the columns to exactly match the project rubric
df_combined = df_combined.rename(columns={'text': 'Text', 'labels': 'Language'})

# 6. Replace the 'en' and 'sw' codes with the full language names
df_combined['Language'] = df_combined['Language'].map({'en': 'English', 'sw': 'Swahili'})

# 7. Shuffle the dataset so that it is a random mix of English and Swahili
df_combined = df_combined.sample(frac=1).reset_index(drop=True)

# 8. Save to CSV
df_combined.to_csv('english_swahili_data.csv', index=False, encoding='utf-8')

print("Success! Clean dataset created and saved to 'english_swahili_data.csv'.\n")

# Verify the rubric requirements:
print(df_combined.head())
print("\n--- Class Balance Check ---")
print(df_combined['Language'].value_counts())
```
