```python
import pandas as pd
import random

print("Initializing Sheng Synthetic Data Generator...")

# 1. Our Matrix of Authentic Sheng Vocabulary
subjects = [
    "Msee", "Mbogi", "Arif", "Manzi", "Huyu boy", 
    "Ma-dem", "Donda", "Karao", "Buda", "Rasta"
]

verbs = [
    "ame-omoka", "ana-dai", "ina-bamba", "ime-weza", "ame-hepa", 
    "ana-zusha", "ame-doz", "ina-lapa", "tuna-chill", "wame-kam"
]

locations_objects = [
    "kwa keja.", "mtaa.", "base.", "chuom.", "tao.", 
    "mashinani.", "kwa matha.", "job.", "downtown.", "area."
]

# 2. Combinatorial Generator
sheng_sentences = set()

# Keep generating until we hit exactly 500 unique sentences
while len(sheng_sentences) < 500:
    # Randomly select one word from each list
    sub = random.choice(subjects)
    verb = random.choice(verbs)
    loc = random.choice(locations_objects)
    
    # Combine them into a 1-2 sentence structure
    sentence = f"{sub} {verb} {loc}"
    sheng_sentences.add(sentence)

# 3. Format to match your Rubric requirements
df_sheng = pd.DataFrame({
    'Text': list(sheng_sentences),
    'Language': 'Sheng'
})

print(f"Success! Synthetically generated {len(df_sheng)} Sheng sentences.")
print(df_sheng.head())

# 4. Save to CSV
df_sheng.to_csv('sheng_data.csv', index=False, encoding='utf-8')
print("\nData saved to 'sheng_data.csv'")
```
