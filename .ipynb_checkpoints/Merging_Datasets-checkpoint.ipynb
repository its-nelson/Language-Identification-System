```python
import pandas as pd

print("Merging datasets...")

# 1. Load the three datasets we created
df_en_sw = pd.read_csv('english_swahili_data.csv')
df_kikuyu = pd.read_csv('kikuyu_data.csv')
df_sheng = pd.read_csv('sheng_data.csv')

# 2. Combine them into one master DataFrame
master_df = pd.concat([df_en_sw, df_kikuyu, df_sheng], ignore_index=True)

# 3. Shuffle the master dataset so the languages are mixed up randomly
# (This is crucial so my ML model does not just memorize the order)
master_df = master_df.sample(frac=1, random_state=42).reset_index(drop=True)

# 4. Save the final dataset for the next phase
master_df.to_csv('master_language_dataset.csv', index=False, encoding='utf-8')

print(f"\nSuccess! Master dataset created with {len(master_df)} rows.")
print("\n--- Final Class Balance ---")
print(master_df['Language'].value_counts())
print("\n--- Sample Data ---")
print(master_df.head())
```
