```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize

# Ensure the tokenizer is ready
nltk.download('punkt')
nltk.download('punkt_tab')

print("Loading Master Dataset...")
# Load the master dataset created in Step 1
df = pd.read_csv('master_language_dataset.csv')

# 1. Custom Slang Handler for Sheng
def handle_sheng_slang(text):
    # This regex finds words connected by a hyphen and merges them into a single word
    # e.g., "ame-doz" becomes "amedoz", "Ma-dem" becomes "Madem"
    text = re.sub(r'(\w+)-(\w+)', r'\1\2', text)
    return text

# 2. The Master Preprocessing Function
def preprocess_text(text):
    if not isinstance(text, str):
        return ""
        
    # Step A: Handle Slang (Must happen BEFORE punctuation removal)
    text = handle_sheng_slang(text)
    
    # Step B: Lowercasing
    text = text.lower()
    
    # Step C: Removing Punctuation (strips out periods, commas, etc.)
    text = re.sub(r'[^\w\s]', '', text)
    
    # Step D: Removing extra spaces (collapses multiple spaces into one)
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Step E: Tokenization (breaks the clean sentence into a list of words)
    tokens = word_tokenize(text)
    
    # Rejoin tokens into a final clean string for the ML models
    return " ".join(tokens)

print("Applying Preprocessing Pipeline...")
# Create a new column with the fully cleaned text
df['Cleaned_Text'] = df['Text'].apply(preprocess_text)

print("\n--- Preprocessing Complete ---")
print(df[['Language', 'Text', 'Cleaned_Text']].head(10))

# Save the preprocessed dataset
df.to_csv('cleaned_master_dataset.csv', index=False, encoding='utf-8')
print("\nCleaned data saved to 'cleaned_master_dataset.csv'!")
```
